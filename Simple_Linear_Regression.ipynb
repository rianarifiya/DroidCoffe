**Modul 1: Menggunakan data sembarang dalam
mengimplementasikan model Regresi Linear**

**Membuat Model Regresi Linear Sederhana**

```python
#import library dan package yang dibutuhkan

import numpy as np #untuk perhitungan saintifik
import matplotlib.pyplot as plt #untuk plotting
from sklearn.linear_model import LinearRegression #import library LinearRegression dari scikit-learn
```

```python

#buat data

penjualan = np.array([6,5,5,4,4,3,2,2,2,1])
harga = np.array([16000, 18000, 27000, 34000, 50000, 68000, 65000, 81000, 85000, 90000])
```

```python
#buat plot
%matplotlib inline
plt.scatter (penjualan, harga)
```

```python
#buat permodelan regresi

penjualan = penjualan.reshape (-1,1) #kita tukar baris dan kolom variabel ini, agar bisa dikalikan dalam operasi matriks
#untuk lebih lengkapnya baca teori soal perhitungan regresi linier

linreg = LinearRegression()
linreg.fit(penjualan, harga)
```

```python

#plot hasil regresi

plt.scatter(penjualan, harga, color='red')
plt.plot(penjualan, linreg.predict(penjualan))
```

Dari analisis yang dilakukan, terlihat adanya hubungan antara harga barang dan jumlah penjualan. Secara umum, ketika harga barang lebih rendah, jumlah penjualan cenderung meningkat. Hal ini menunjukkan bahwa harga memiliki pengaruh terhadap minat beli konsumen.

**Bermain dengan Dataset**

```python
#import library dan package yang dibutuhkan

import pandas as pd #untuk dataframe
import pylab as pl #untuk plotting
%matplotlib inline
```

```python
#download data

!wget -O FuelConsumption.csv https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/FuelConsumptionCo2.csv
```

Dataset ini berisi data konsumsi bahan bakar dan estimasi emisi CO₂ pada beberapa kendaraan yang dijual di Canada. Dataset ini berisi data :
- MODELYEAR
- MAKE
- MODEL
- VEHICLE CLASS
- ENGINE SIZE
- CYLINDERS
- TRANSMISSION
- FUEL CONSUMTION in CITY (L/100 km)
- FUEL CONSUMTION in HWY (L/100 km)
- FUEL CONSUMTION COMB (L/100 km)
- CO₂ EMISSIONS (g/km)

```python
df = pd.read_csv("FuelConsumption.csv") #membaca data

# melihat 5 baris pertama data
df.head()
```

```python
#kita ambil kolom mana saja yang akan kita analisis, dan membuang sisanya

cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','CO2EMISSIONS']]
cdf.head(9)
```

```python
#Plot hubungan konsumsi bahan bakar dengan emisi CO₂

plt.scatter(cdf.FUELCONSUMPTION_CITY, cdf.CO2EMISSIONS,  color='blue')
plt.xlabel("FUELCONSUMPTION_CITY")
plt.ylabel("Emission")
plt.show()
```

Diagram ini menunjukkan hubungan antara konsumsi bahan bakar dengan emisi CO₂.

```python
#Plot hubungan ukuran mesin dengan emisi CO₂

plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

Hubungan antara variabel ukuran mesin dengan emisi CO₂ dapat kita lihat melalui diagram di atas.

**Melakukan Data Splitting**

```python
#Membagi data

msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

Pada bagian ini, kita akan mencari hubungan antara ukuran mesin dengan emisi CO₂ pada kendaraan.

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

**Membuat Model Regresi**

```python
#Membuat model regresi
regr = LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (train_x, train_y)

# Koefisien model
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

Dalam membangun model regresi, pada dasarnya kita mencari pola hubungan antara variabel input dan hasil yang ingin diprediksi. Pola ini direpresentasikan dalam bentuk garis miring atau _coefficient_, di mana kemiringan garis menunjukkan seberapa besar pengaruh suatu faktor, dan titik awalnya menunjukkan nilai dasar ketika tidak ada pengaruh dari faktor tersebut. Memahami konsep ini akan membantu dalam melihat bagaimana model menghasilkan prediksi.

```python
#Plot hasil regresi

plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
```

```python
#Menghitung error

from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
test_y_ = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_ , test_y) )
```

Dari hasil yang diperoleh, terlihat bahwa tingkat kesalahan prediksi model masih cukup besar. Hal ini terjadi karena emisi CO₂ akan tetapi tidak hanya dipengaruhi oleh satu faktor saja, tetapi juga oleh berbagai faktor lain yang belum dimasukkan dalam model. Model yang digunakan saat ini masih bersifat sederhana, sehingga kemampuannya dalam menangkap kompleksitas data masih terbatas. Oleh karena itu, diperlukan pendekatan yang lebih komprehensif agar hasil prediksi dapat menjadi lebih akurat.
